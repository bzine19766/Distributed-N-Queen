```python
#number of  edges mathematically 
for n in range(5,12):
    m=n-1
    sm =m*m
    ne= (sm*(sm-1))//2
    ne=(ne-((m)*(m-1)*(5*m-1))//3)
    print("The number of edges in ",m ,"queen =", ne)
```
