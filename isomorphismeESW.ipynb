```python
G = nx.Graph()
```

```python
Dict = {} 
```

```python
for i in range(nn):
    G.add_node(i)
    Dict[lnn[i]]=i
```

```python
for i in lnn:
    for j in GF.adj[i]:
        G.add_edge(Dict[i],Dict[j])
```
