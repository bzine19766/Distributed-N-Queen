```python
for i in range(1,n):
    for j in range(1,n):
        for k in range(1,n):
            for l in range(1,n):
                if (i!=k and j!=l) and ((i+j!=k+l) and (i-j!=k-l)) :
                        #GF.add_edge (int(str(i)+str(j)),int(str(k)+str(l)))
                        GF.add_edge ((i,j),(k,l))
```
